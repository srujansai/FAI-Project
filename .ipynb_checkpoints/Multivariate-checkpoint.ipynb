```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as stap
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression  
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
%matplotlib inline
```

```python
#read, initialize and clean data
t_df = pd.read_csv('temperature_data.csv', usecols=[2,3,4,5,7,8,11,12,13], engine='python')
t_df['YEARMODA'] = pd.to_datetime(t_df['YEARMODA'], format='%Y%m%d')
t_df['PRCP'] = t_df['PRCP'].astype(str).str[:-1].astype(np.float64)
t_vars = t_df.columns.values.tolist()[1:]
# print(t_df)
# print(t_vars)


#Replace outliers/missing data with mean or 0 based on the variable

def exclude(lst, i):
    if i == 0:
        return lst[i+1:]
    return lst[:i] + lst[i+1:]

#for mean, max, min tmp's and SLP
def fill_with_mask1(s):
    mean = round(s.loc[lambda x: x != 0].mean(), 1)
    return s.mask(s.eq(9999.9), mean)

for k in [0,1,2,5,6]:
    gidx = exclude(t_vars,k)
    t_df[t_vars[k]] = t_df.groupby(gidx)[t_vars[k]].apply(fill_with_mask1)

#for visibilty and wind speed
def fill_with_mask2(s):
    mean = round(s.loc[lambda x: x != 0].mean(), 1)
    return s.mask(s.eq(999.9), mean)

for k in [3,4]:
    gidx = exclude(t_vars,k)
    t_df[t_vars[k]] = t_df.groupby(gidx)[t_vars[k]].apply(fill_with_mask2)
    
#for precipitation
def fill_with_mask3(s):
    return s.mask(s.eq(99.99), 0.00)

gidx = exclude(t_vars,7)
t_df[t_vars[7]] = t_df.groupby(gidx)[t_vars[7]].apply(fill_with_mask3)


""" create a 2 day window for the variables"""
def n_window(df, variable, n):  
    rows = t_df.shape[0]
    n_window_val = [None]*n + [df[variable][i-n] for i in range(n, rows)]
    col_name = "{}_{}".format(variable, n)
    df[col_name] = n_window_val
    
    
for t_var in t_vars:  
    for n in range(1, 3):
        n_window(t_df, t_var, n)
        
        
#Keep only the dependant variable at t and all independent varaibles at t-1 and t-2        
t_df = t_df[[col for col in t_df.columns if col not in t_vars[1:]]] 
t_df = t_df.dropna() 
t_df.info()  
```

# Linear Regression

```python
# dependent and independent variables seperated
total_vars = t_df.columns.values.tolist()[1:]
lm_df = t_df[total_vars]
inde_vars = total_vars[1:]
X = lm_df[inde_vars]  
y = lm_df[total_vars[:1]]

# Add a constant to the independent variable set to represent the beta0 intercept
X = stap.add_constant(X)  
X.ix[:5, :5] 
print(X)
print(total_vars[:1])
```

```python
min_sig = 0.005

#iter2
X = X.drop('DEWP_1', axis=1)
#iter3
X = X.drop('DEWP_2', axis=1)
#iter4
X = X.drop('SLP_2', axis=1)
#iter5
X = X.drop('MAX_2', axis=1)
#iter6
X = X.drop('PRCP_2', axis=1)
#iter7
X = X.drop('PRCP_1', axis=1)
#iter8
X = X.drop('SLP_1', axis=1)
#iter9
X = X.drop('VISIB_1', axis=1)


# Fit the model
model = stap.OLS(y, X).fit()

# model stats
model.summary() 
```

```python
# Based on the above arrived at features build a regressor
X = X.drop('const', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.31, random_state=9) 
```

```python
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)

# make a prediction set using the test set
prediction = regression_model.predict(X_test)

# Evaluate the prediction accuracy of the model 
print("The Explained Variance train: %.4f" % regression_model.score(X_train, y_train))  
print("The Explained Variance test: %.4f" % regression_model.score(X_test, y_test))  
print("The Root Mean Squared Error test: %.2f degrees Fahrenheit" % np.sqrt(mean_squared_error(y_test, prediction)))
print("The R squared: %.4f" % r2_score(y_test, prediction))
```

## DNN Regressor

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.metrics import explained_variance_score, \
    mean_squared_error, \
    mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
# dependent and independent variables seperated
total_vars = t_df.columns.values.tolist()[1:]
lm_df = t_df[total_vars]
inde_vars = total_vars[1:]
X = lm_df[inde_vars]  
y = lm_df[total_vars[:1]]
```

```python
#Get training data
X_train, X_remain, y_train, y_remain = train_test_split(X, y, test_size=0.31, random_state=9)
#Set testing and Validation data
X_val, X_test, y_val, y_test = train_test_split(X_remain, y_remain, test_size=0.5, random_state=9)
```

```python
#feature initialization
features = [tf.feature_column.numeric_column(xcolumn) for xcolumn in X.columns]

#regressor initialization
regressor = tf.estimator.DNNRegressor(feature_columns=features,
                                      hidden_units=[81, 40, 9],
                                      model_dir='tf_fai_model')
```

```python
def fai_in_fn(X, y=None, num_epochs=None, shuffle=True, batch_size=300):
    return tf.estimator.inputs.pandas_input_fn(x=X,
                                               y=y,
                                               num_epochs=num_epochs,
                                               shuffle=shuffle,
                                               batch_size=batch_size)
```

```python
evals = []
STEPS = 500
for i in range(100):
    regressor.train(input_fn=fai_in_fn(X_train, y=y_train), steps=STEPS)
    evaluation = regressor.evaluate(input_fn=fai_in_fn(X_val, y_val,num_epochs=1,shuffle=False),steps=1)
    evals.append(regressor.evaluate(input_fn=fai_in_fn(X_val,y_val,num_epochs=1,shuffle=False)))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# manually set the parameters of the figure to and appropriate size
plt.figure(figsize=(15,6),dpi=150)

#plt.rcParams['figure.figsize'] = [14, 10]

loss_values = [ev['loss'] for ev in evals]
training_steps = [ev['global_step'] for ev in evals]

plt.scatter(x=training_steps, y=loss_values)
plt.xlabel('Training steps (Epochs)')
plt.ylabel('Loss (SSE)')
plt.tight_layout()
plt.savefig("epoch.pdf")
plt.show()
```

```python
#Forecast with testing data
pred = regressor.predict(input_fn=fai_in_fn(X_test,num_epochs=1,shuffle=False))
predictions = np.array([p['predictions'][0] for p in pred])

print("The Explained Variance: %.2f" % explained_variance_score(
                                            y_test, predictions))  
print("The Mean Absolute Error test: %.2f degrees Fahrenheit" % mean_absolute_error(
                                            y_test, predictions))  
print("The Root Mean Squared Error test: %.2f degrees Fahrenheit" % np.sqrt(mean_squared_error(
                                            y_test, predictions)))

print("The R squared test: %.3f " % r2_score(
                                            y_test, predictions))
```
